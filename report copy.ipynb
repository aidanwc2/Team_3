# How does Tobacco Usage Vary Between Different Regions of the World and What is the Reasoning Behind This?


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("tobacco_use_ww.csv")
df
```

### Seperating Data by Region

```python
df_Asia = df[df["ParentLocation"] == "South-East Asia"]
df_Africa = df[df["ParentLocation"] == "Africa"]
df_Europe = df[df["ParentLocation"] == "Europe"]
df_EastMed = df[df["ParentLocation"] == "Eastern Mediterranean"]
df_WestPac = df[df["ParentLocation"] == "Western Pacific"]
df_Americas = df[df["ParentLocation"] == "Americas"]
```

```python
df_AmericasM = df_Americas[df_Americas["Gender"] == "Male"]
df_AmericasF = df_Americas[df_Americas["Gender"] == "Female"]

print(df_AmericasM["Value"].hist(density = True, color = 'forestgreen'))
plt.title('Tobacco Usage in the Americas (Males)', fontsize=20)
plt.xlabel('Tobacco Usage (%)')
plt.ylabel('Density')
plt.figure()

print(df_AmericasF["Value"].hist(density = True, color = 'forestgreen'))
plt.title('Tobacco Usage in the Americas (Females)', fontsize=20)
plt.xlabel('Tobacco Usage (%)')
plt.ylabel('Density')
plt.figure()

print(df_AmericasM["Value"].mean())
print(df_AmericasM["Value"].std())
print(df_AmericasF["Value"].mean())
print(df_AmericasF["Value"].std())
```

```python
df_AfricaM = df_Africa[df_Africa["Gender"] == "Male"]
df_AfricaF = df_Africa[df_Africa["Gender"] == "Female"]

print(df_AfricaM["Value"].hist(density = True,color = 'sandybrown'))
plt.title('Tobacco Usage in Africa (Males)', fontsize=20)
plt.xlabel('Tobacco Usage (%)')
plt.ylabel('Density')
plt.figure()
print(df_AfricaF["Value"].hist(density = True,color = 'sandybrown'))
plt.title('Tobacco Usage in Africa (Females)', fontsize=20)
plt.xlabel('Tobacco Usage (%)')
plt.ylabel('Density')
plt.figure()

print(df_AfricaM["Value"].mean())
print(df_AfricaM["Value"].std())
print(df_AfricaF["Value"].mean())
print(df_AfricaF["Value"].std())
```

```python
df_AsiaM = df_Asia[df_Asia["Gender"] == "Male"]
df_AsiaF = df_Asia[df_Asia["Gender"] == "Female"]

print(df_AsiaM["Value"].hist(density = True,color = 'indianred'))
plt.title('Tobacco Usage in South-East Asia (Males)', fontsize=20)
plt.xlabel('Tobacco Usage (%)')
plt.ylabel('Density')
plt.figure()
print(df_AsiaF["Value"].hist(density = True,color = 'indianred'))
plt.title('Tobacco Usage in South-East Asia (Females)', fontsize=20)
plt.xlabel('Tobacco Usage (%)')
plt.ylabel('Density')
plt.figure()

print(df_AsiaM["Value"].mean())
print(df_AsiaM["Value"].std())
print(df_AsiaF["Value"].mean())
print(df_AsiaF["Value"].std())
```

```python
df_WestPacM = df_WestPac[df_WestPac["Gender"] == "Male"]
df_WestPacF = df_WestPac[df_WestPac["Gender"] == "Female"]

print(df_WestPacM["Value"].hist(density = True,color = 'olive'))
plt.title('Tobacco Usage in the Western Pacific (Males)', fontsize=20)
plt.xlabel('Tobacco Usage (%)')
plt.ylabel('Density')
plt.figure()
print(df_WestPacF["Value"].hist(density = True,color = 'olive'))
plt.title('Tobacco Usage in the Western Pacific (Females)', fontsize=20)
plt.xlabel('Tobacco Usage (%)')
plt.ylabel('Density')
plt.figure()

print(df_WestPacM["Value"].mean())
print(df_WestPacM["Value"].std())
print(df_WestPacF["Value"].mean())
print(df_WestPacF["Value"].std())
```

### Relationship Between Usage Value and Average Cigarette Price For Each Region 

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python

df_combined = pd.read_csv("CombinedDataMaleFinal.csv")
x = np.array(df_combined["Value"])
y = np.array(df_combined["AvgCigarettePriceDollars"])


df_combined_q = df_combined.dropna()

X = df_combined_q['Value'].values[:,np.newaxis]
y = df_combined_q['AvgCigarettePriceDollars'].values


model = LinearRegression()
model.fit(X,y)
plt.scatter(X, y)
plt.plot(X, model.predict(X),color='k')
plt.title('Average Price of Cigarette vs Tobacco Usage (Males)', fontsize=20)
plt.xlabel('Tobacco Usage (%)')
plt.ylabel('Average Price of Cigarette ($)')
plt.show()

print(model.coef_)
print(model.intercept_)
df_combined.corr()
```
