# How does Tobacco Usage Differ Between Different Regions of the World and What is the Reasoning Behind This?


```python
import pandas as pd
df = pd.read_csv("tobacco_use_ww.csv")
df
```

### Seperating Data by Region

```python
df_Asia = df[df["ParentLocation"] == "South-East Asia"]
df_Asia["Value"].mean()
```

```python
df_Africa = df[df["ParentLocation"] == "Africa"]
df_Africa["Value"].mean()
```

```python
df_Europe = df[df["ParentLocation"] == "Europe"]
df_Europe["Value"].mean()
```

```python
df_Africa = df[df["ParentLocation"] == "Africa"]
df_Africa["Value"].mean()
```

```python
df_EastMed = df[df["ParentLocation"] == "Eastern Mediterranean"]
df_EastMed["Value"].mean()
```

```python
df_WestPac = df[df["ParentLocation"] == "Western Pacific"]
df_WestPac["Value"].mean()
```

```python
df_Americas = df[df["ParentLocation"] == "Americas"]
df_Americas["Value"].mean()
```

```python
len(df_Africa) + len(df_WestPac) + len(df_Asia) + len(df_EastMed) + len(df_Europe) + len(df_Americas)
```

```python
import matplotlib.pyplot as plt
print(df_Americas["Value"].hist(color = 'forestgreen'))
plt.title('Tobacco Usage in the Americas', fontsize=20)
plt.xlabel('Usage')
plt.ylabel('Count')
plt.figure()

print(df_Africa["Value"].hist(color = 'sandybrown'))
plt.title('Tobacco Usage in Africa', fontsize=20)
plt.xlabel('Usage')
plt.ylabel('Count')
plt.figure()

print(df_Asia["Value"].hist(color = 'indianred'))
plt.title('Tobacco Usage in Asia', fontsize=20)
plt.xlabel('Usage')
plt.ylabel('Count')
plt.figure()

print(df_EastMed["Value"].hist(color = 'navy'))
plt.title('Tobacco Usage in the Eastern Mediterranean', fontsize=20)
plt.xlabel('Usage')
plt.ylabel('Count')
plt.figure()

print(df_WestPac["Value"].hist(color = 'olive'))
plt.title('Tobacco Usage in the Western Pacific', fontsize=20)
plt.xlabel('Usage')
plt.ylabel('Count')
plt.figure()

print(df_Europe["Value"].hist(color = 'powderblue'))
plt.title('Tobacco Usage in Europe', fontsize=20)
plt.xlabel('Usage')
plt.ylabel('Count')
plt.figure()
```

```python
import numpy as np
df_combined = pd.read_csv("CombinedDataMale.csv")
x = np.array(df_combined["Value"])
y = np.array(df_combined["AvgCigarettePriceDollars"])
a, b = np.polyfit(x,y,1)

df_combined.plot.scatter(x='Value', y="AvgCigarettePriceDollars")
plt.plot(x, a*x + b)
```
